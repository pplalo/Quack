```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ruta = "web-edu.mtx"

datos = pd.read_csv(ruta, skiprows=2, sep=" ", header=None)

G = nx.from_pandas_edgelist(datos,0,1)
```

## Número de enlaces y nodos

```python
len(G.nodes), len(G.edges)
```

## Clustering promedio

```python
C_max = max(nx.connected_components(G), key = len)
Gn = nx.subgraph(G, C_max)
```

```python
2*len(Gn.edges)/len(Gn)
```

## Longitud de camino promedio

```python
nx.average_shortest_path_length(Gn)
```

## Diámetro

```python
nx.diameter(Gn)
```

## Promedio del clustering

```python
nx.average_clustering(Gn)
```

## Distribución de grado

```python
dist = {nx.degree(Gn)[nodo]:0 for nodo in Gn}
```

```python
for nodo in Gn:
    dist[nx.degree(Gn)[nodo]] += 1

X = dist.keys()
Y = dist.values()

plt.plot(X,Y,"o")
```

```python
dist
```

```python
plt.bar(X,Y)
plt.xlabel('Grados')
plt.ylabel('Número de grados')
plt.title('Grado de los nodos')
plt.xticks(ticks = range(0,max(X)+1,5),labels = range(0,max(X)+1,5))
plt.show()
```
