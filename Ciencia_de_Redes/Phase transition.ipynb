```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.erdos_renyi_graph(50,.05)
nx.draw(G, with_labels=True)
```

```python
list(nx.connected_components(G))
```

```python
len(max(nx.connected_components(G), key=len))
```

Notamos que los valores asignados a N al momento de la normalización mientras más grande sea N más se acerca a 1 y viceversa.

```python
N = 100
G = nx.erdos_renyi_graph(N,.05)
nx.draw(G, with_labels=True,)

len(max(nx.connected_components(G), key=len))/N
```

Generamos un dominio

```python
N = 100
dominio = np.linspace(0,1,100)
for p in dominio:
    G = nx.erdos_renyi_graph(N,p)
    print(len(max(nx.connected_components(G), key=len))/N)


```
