```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

file = open("econ-mbeacxc.mtx", "r")
lineas = file.readlines()
file.close()
```

```python
G = nx.Graph()

for linea in lineas:
    if linea[0] == "%": continue
    else:
       n1, n2, w =  linea[:-1].split(" ")
       n1, n2, w = int(n1), int(n2), float(w)
       if n1 == n2: continue
       G.add_edge(n1, n2)
    #    G.add_edge(n1, n2, weight=w)
```

```python
# print(G.edges(data=True))
pos = nx.kamada_kawai_layout(G)
```

```python
nx.draw(G, pos, node_size = 1, width = 0.1, edge_color = "gray")
```

```python
# Al tamao de cada nodo le asignamos el grado del nodo para que visualmente se vea.
tamaos = [G.degree(n) for n in G.nodes()]
nx.draw(G, pos, node_size = tamaos, edgecolors = "b", width = 0.1, node_color = tamaos, cmap = "plasma")
```

```python
pos = nx.circular_layout(G)
nx.draw_networkx_edges(G, pos, width = 0.1, edge_color = "gray",alpha = 0.05)
```
