##Resolución del oscilador armónico simple


```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Aquí tenemos todos los parámetros que vamos a usar incluyendo las cosntantes de oscilación y demás parámetros
k=8.0
l=0.5
m=1.0
b=0.4
g=b/(2*m**2)
w=np.sqrt(k/m-(b**2)/(4*m**2))
# Las condiciones iniciales se hicieron por separado debido a que hubo muchos problemas en la ejecución tratando de usar arrays
x0=1
v0=0

# Se definieron 2 funciones donde cada una de éstas describiría el conjunto solución tanto para la velocidad como el tiempo de nuestro 
# oscilador armónico amortiguado

# Este fue para la posición
def f(x,t):
  return np.exp(-g*t)*(np.cos(w*t) + b/((2*m**2)*(w))*np.sin(w*t))
# Este fue para la velocidad
def h(x,t):
  return -g*np.exp(-g*t)*(np.cos(w*t)+b/((2*m**2)*(w))*np.sin(w*t))+np.exp(-g*t)*(-w*np.sin(w*t)+b/((2*m**2))*np.cos(w*t))

t=np.arange(0,20.1,0.01)  # Los puntos en t donde la función se iba a evaluar
x=odeint(f,x0,t) # La función a describir para la posición
v=odeint(h,v0,t) # La función a describir para la velocidad
```

```python
# Grafica de tiempo contra posición
fig,ax = plt.subplots()
ax.plot(t,x,'k')
ax.set_xlabel('Tiempo[s]')
ax.set_title('Oscilador amortiguado')
ax.set_ylabel('x(t)')
fig.tight_layout()
plt.show()
```

```python
# Grafica del tiempo contra la velocidad
fig,ax = plt.subplots(figsize=(7,8))
ax.plot(t,v,'y')
ax.set_xlabel('Tiempo[s]')
ax.set_title('Oscilador amortiguado')
ax.set_ylabel('v(t)')
fig.tight_layout()
plt.show()
```

```python
# Para el espacio fase donde graficamos posición y tiempo
fig,ax = plt.subplots(figsize=(7,8))
ax.plot(x,v,'b')
ax.set_xlabel('Posición')
ax.set_title('Espacio Fase')
ax.set_ylabel('Velocidad')
fig.tight_layout()
plt.show()
```
