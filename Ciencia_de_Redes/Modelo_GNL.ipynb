Se construyen las funciones para generar redes aleatorias mediante el modelo G(N,L) es decir fijando nodos y número de enlaces

```python
import networkx as nx
import numpy as np
```

```python
N  = 100 # Número de nodos
L = 200 # Número de enlaces
```

```python
np.random.choice(range(N), size=2, replace=False)
```

```python
def GNL(N,L):
    G = nx.Graph()
    A = np.random.choice(N,[L,2])
    A = A[A[:,0] != A[:,1]]
    G.add_edges_from(A)

    while len(G.edges) < L:
        e = np.random.choice(N,2)
        if e[0] == e[1]: continue
        G.add_edge(e[0],e[1])
    return G
```
