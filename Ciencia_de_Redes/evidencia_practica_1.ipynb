```python
import networkx as nw

G = nw.Graph()
G.add_node(37)

for i in range(37):
    G.add_edge(1, i+2)

nw.draw(G, with_labels = True, node_size = 1000, node_color = 'red')
```

```python
G = nw.Graph()
G.add_node(42)

for i in range(42):
    for j in range(42):
        G.add_edge(1, j+2)
        G.add_edge(2, j+3)

nw.draw(G, node_size = 10, node_color = 'green')
```
